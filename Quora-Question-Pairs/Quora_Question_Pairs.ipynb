# Feature Extraction

```python
def WordsCounter(data):
  data['q1_len']=data['question1'].str.len()  # Finding length of Question 1
  data['q2_len']=data['question2'].str.len()  # Finding length of Question 2

  data['q1_words']=data['question1'].apply(lambda row : len(row.split()))   # Finding Number of Words in Question 1
  data['q2_words']=data['question2'].apply(lambda row : len(row.split()))   # Finding Number of Words in Question 2
```

```python
def WordShare(row):

    w1 = set(map(lambda word: word.lower().strip(), row['question1'].split(" ")))
    w2 = set(map(lambda word: word.lower().strip(), row['question2'].split(" ")))    
    return 1.0 * len(w1 & w2)/(len(w1) + len(w2))    # Finding the number of common words between question1 and question2 and dividing by total words between both of them
```

```python
def WordShareRate(row):
    w1 = set(map(lambda word: word.lower().strip(), row['question1'].split(" ")))
    w2 = set(map(lambda word: word.lower().strip(), row['question2'].split(" ")))    
    return 1.0 * len(w1 & w2)/(len(w1) + len(w2))    # Finding the number of common words between question1 and question2 and dividing by total words between both of them
```

```python
def WordShare(row):
    w1 = set(map(lambda word: word.lower().strip(), row['question1'].split(" ")))
    w2 = set(map(lambda word: word.lower().strip(), row['question2'].split(" ")))    
    return len(w1 & w2) 
```

```python
def FeatureExtraction(data):
   WordsCounter(data)
   data['question1'] = data['question1'].astype(str)
   data['question2'] = data['question2'].astype(str)
   data['Shared_Word'] = data.apply(WordShare, axis=1)
   data['Shared_Word_Rate'] = data.apply(WordShareRate,axis=1)
```

```python
punct = ['/','[',']','"',"'",'{','}',"\\" ]
exceptions =  ['-','.',"+","=","*",":"]
```

```python
import string 
# Create a function for removing punctuation
def RemovePunctaution(text):
    repacePunctaution = []
    for words in text:
      if words in punct:
        repacePunctaution.append(' ')
      else:
        repacePunctaution.append(words)
    noPunctaution = [words for words in ''.join(repacePunctaution) if ( words not in string.punctuation or words in exceptions)]
    return ' '.join(''.join(noPunctaution).split())
```

```python
import string 
import re
# Create a function for removing punctuation
def ReplacePunctaution(text):
    words = text.split()
    noPunctaution = []
    for word in words:
       if word in abbrDict.keys():
            noPunctaution.append(abbrDict[word])
       else :
            noPunctaution.append(word)
    return ' '.join(noPunctaution)
```

```python
# Make a fucntion for convert Upper letter to lower letter and number remover
def UpperToLower(text):
    noUpperCase = [words.lower() for words in text]
    return ''.join(noUpperCase)
```
