```python
def WordsCounter(data):
  data['q1_len']=data['question1'].str.len()  # Finding length of Question 1
  data['q2_len']=data['question2'].str.len()  # Finding length of Question 2

  data['q1_words']=data['question1'].apply(lambda row : len(row.split()))   # Finding Number of Words in Question 1
  data['q2_words']=data['question2'].apply(lambda row : len(row.split()))   # Finding Number of Words in Question 2
```

```python
def WordShare(row):

    w1 = set(map(lambda word: word.lower().strip(), row['question1'].split(" ")))
    w2 = set(map(lambda word: word.lower().strip(), row['question2'].split(" ")))    
    return 1.0 * len(w1 & w2)/(len(w1) + len(w2))    # Finding the number of common words between question1 and question2 and dividing by total words between both of them
```

```python
def WordShareRate(row):
    w1 = set(map(lambda word: word.lower().strip(), row['question1'].split(" ")))
    w2 = set(map(lambda word: word.lower().strip(), row['question2'].split(" ")))    
    return 1.0 * len(w1 & w2)/(len(w1) + len(w2))    # Finding the number of common words between question1 and question2 and dividing by total words between both of them
```

```python
def WordShare(row):
    w1 = set(map(lambda word: word.lower().strip(), row['question1'].split(" ")))
    w2 = set(map(lambda word: word.lower().strip(), row['question2'].split(" ")))    
    return len(w1 & w2) 
```

```python
def FeatureExtraction(data):
   data['question1'] = data['question1'].astype(str)
   data['question2'] = data['question2'].astype(str)
   WordsCounter(data)
   data['Shared_Word'] = data.apply(WordShare, axis=1)
   data['Shared_Word_Rate'] = data.apply(WordShareRate,axis=1)
```

```python
punct = ['/','[',']','"',"'",'{','}',"\\" ]
exceptions =  ['-','.',"+","=","*",":"]
```

```python
import string 
# Create a function for removing punctuation
def RemovePunctaution(text):
    repacePunctaution = []
    for words in text:
      if words in punct:
        repacePunctaution.append(' ')
      else:
        repacePunctaution.append(words)
    noPunctaution = [words for words in ''.join(repacePunctaution) if ( words not in string.punctuation or words in exceptions)]
    return ' '.join(''.join(noPunctaution).split())
```

```python
import string 
import re
# Create a function for removing punctuation
def ReplacePunctaution(text):
    words = text.split()
    noPunctaution = []
    for word in words:
       if word in abbrDict.keys():
            noPunctaution.append(abbrDict[word])
       else :
            noPunctaution.append(word)
    return ' '.join(noPunctaution)
```

```python
# Make a fucntion for convert Upper letter to lower letter and number remover
def UpperToLower(text):
    noUpperCase = [words.lower() for words in text]
    return ''.join(noUpperCase)
```

```python
# TODO : use stopwords after building model and  remove number  adn lemmitize
def Pipline(data):
  print(f'shpae before pipline {data.shape}')
  data.drop(data.index[rowsWithNan])
  for question in ['question1','question2']:
    data[question] = data[question].astype(str)
    data[f'{question}-after-cleaning'] = data[question].apply(lambda x : UpperToLower(x))
    data[f'{question}-after-cleaning'] = data[f'{question}-after-cleaning'].apply(lambda x : ReplacePunctaution(x))
    data[f'{question}-after-cleaning'] = data[f'{question}-after-cleaning'].apply(lambda x : RemovePunctaution(x))
  print(f'shpae after pipline {data.shape}')

```

```python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
from collections import Counter

train_data['Simple_Ratio']=train_data.apply(lambda x: fuzz.ratio(x['question1-after-cleaning'],x['question2-after-cleaning']) ,axis=1)

train_data['Partial_Ratio']=train_data.apply(lambda x: fuzz.partial_ratio(x['question1-after-cleaning'],x['question2-after-cleaning']) ,axis=1)

train_data['Token_Sort_Ratio']=train_data.apply(lambda x: fuzz.token_sort_ratio(x['question1-after-cleaning'],x['question2-after-cleaning']) ,axis=1)

train_data['Token_Set_Ratio']=train_data.apply(lambda x: fuzz.token_set_ratio(x['question1-after-cleaning'],x['question2-after-cleaning']) ,axis=1)
train_data['Last_Word']=train_data.apply(lambda x: int(x['question1'].split()[-1] == x['question2'].split()[-1]),axis=1)


train_data['First_Word']=train_data.apply(lambda x: int(x['question1'].split()[0] == x['question2'].split()[0]),axis=1)


train_data['Length_diff']=train_data.apply(lambda x: abs(len(x['question1'].split())-len(x['question2'].split())),axis=1)
def commonStopWordRatio(q1,q2):
    q1_token=q1.split() # Splitting Words to make then tokens
    q2_token=q2.split()
    
    # We are takking out all the stopwords in both the Question and finding there intersection (Common stopwords)
    q1=set([word for word in q1_token if word in stopwords])
    q2=set([word for word in q2_token if word in stopwords])
    common=q1.intersection(q2)
    ratio=len(common)/len(set(q1_token))+len(set(q2_token))
    
    return ratio

def CommonTokenRatio(q1,q2):
    q1Token=q1.split()
    q2Token=q2.split()
    # We are takking out all the token in both the Question and finding there intersection (Common tokens)
    q1=set([x for x in q1Token if x not in stopwords])
    q2=set([x for x in q2Token if x not in stopwords])
    
    common=q1.intersection(q2)
    ratio=len(common)/len(set(q1Token))+len(set(q2Token))
    
    return ratio



def CommonCharacters(q1,q2):
    q1="".join(q1.split())
    q2="".join(q2.split())
    c=0
    d=dict(Counter(q1))
    for i in q2:
        if i in d and d[i]!=0:
            c+=1;d[i]-=1
    return c


train_data['StopWord_Ratio']=train_data.apply(lambda x: commonStopWordRatio(x["question1"], x["question2"]), axis=1)


train_data['Token_Ratio']=train_data.apply(lambda x: CommonTokenRatio(x["question1"], x["question2"]), axis=1)


```
